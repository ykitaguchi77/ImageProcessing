```python
from pdf2image import convert_from_path
import tempfile
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline 


with tempfile.TemporaryDirectory() as path:
    img = convert_from_path("/content/56cm.pdf", output_folder=path)

orig = img[0]
dst = orig.resize((2008, 1500), Image.BILINEAR)

print(f"{orig.size} -> {dst.size}")

arrPIL = np.asarray(dst)
plt.imshow(arrPIL)
plt.show()


_dpi = 300
dst.save("/content/img0.tiff", dpi = (_dpi, _dpi)) # 新しいDPIに書き換えて出力する
dst.close() # 閉じる
```
