```python
"""
666mai---grav
       |
       |-cont

gravcont_all_500px---grav
                   |
                   |-cont

↓               

dataset---train----grav
        |       |--cont
        |
        |-val------grav  
        |       |--cont
        |
        |-test------grav  
                |--cont

"""

import os
import shutil
import random
import time
```

#**フォルダの指定**

```python
#評価済666maiデータセット
handai_path = "/content/drive/MyDrive/Deep_learning/666mai_dataset/gravcont_500px"

#オリンピアgrav＋阪大データセット（上記を含んでいる）
olympia_path = "/content/drive/MyDrive/Deep_learning/gravcont_all_500px"

#新しくデータを作成する場所
dst_path = "/content/drive/MyDrive/Deep_learning/gravcont_all_500px_divided"

#trainとvalの割合を指定。残りがtestになる
rate_train = 0.74
rate_val = 0.185
```

#**各クラスのデータ数カウント**

```python
#親フォルダ直下のディレクトリを探索
files = os.listdir(handai_path)
files_dir = [f for f in files if os.path.isdir(os.path.join(handai_path, f))]
print(files_dir)

#666枚フォルダのリスト
handai_grav = os.listdir(handai_path+"/grav")
handai_cont = os.listdir(handai_path+"/cont")

#allフォルダのリスト
olympia_grav = os.listdir(olympia_path+"/grav")
olympia_cont = os.listdir(olympia_path+"/cont")

#各フォルダの合計（重複除く）
total_grav = list(set(handai_grav + olympia_grav))
total_cont = list(set(handai_cont + olympia_cont))
total = list(set(total_grav + total_cont))

print("grav: "+str(len(total_grav)))
print("cont: "+str(len(total_cont)))

#それぞれの項目の枚数を算出
grav_train_num = int(len(total_grav)*rate_train)
grav_val_num = int(len(total_grav)*rate_val)
grav_test_num = len(total_grav)-grav_train_num-grav_val_num
cont_train_num = int(len(total_cont)*rate_train)
cont_val_num = int(len(total_cont)*rate_val)
cont_test_num = len(total_cont)-cont_train_num-cont_val_num

print("")
print("【grav】")
print("train "+str(grav_train_num))
print("val "+str(grav_val_num))
print("test "+str(grav_test_num))

print("")
print("【cont】")
print("train "+str(cont_train_num))
print("val "+str(cont_val_num))
print("test "+str(cont_test_num))
```

```python
#testsetはhandaiフォルダから作る
test_grav = random.sample(handai_grav, grav_test_num)
test_cont = random.sample(handai_cont, cont_test_num)

#残りからtrainsetとvalsetを作る
train_grav = random.sample(list(set(total_grav)-set(test_grav)), grav_train_num)
train_cont = random.sample(list(set(total_cont)-set(test_cont)), cont_train_num)
val_grav = list(set(total_grav)-set(test_grav)-set(train_grav))
val_cont = list(set(total_cont)-set(test_cont)-set(train_cont))
```

#**空フォルダの作成**

```python
#目的フォルダがあったら初期化
if os.path.exists(dst_path):
    shutil.rmtree(dst_path)
os.makedirs(dst_path, exist_ok=True)

#train, val, testフォルダ、および対応する子フォルダを作成
for i in files_dir:
    os.makedirs(dst_path +'/train/' + i, exist_ok=True)
    os.makedirs(dst_path +'/val/' + i, exist_ok=True)
    os.makedirs(dst_path +'/test/' + i, exist_ok=True)
```

```python
#ファイル移動

files = [test_grav, test_cont, train_grav, train_cont, val_grav, val_cont]
dst_phase = ["/test/grav/", "/test/cont/", "/train/grav/", "/train/cont/", "/val/grav/", "/val/cont/"]

start = time.time()

k=0
l=0
for i in files:
    for j in i:
        if os.path.exists(olympia_path+"/grav/"+str(j)):
            shutil.copy(olympia_path+"/grav/"+str(j), dst_path+dst_phase[k] + str(j))
        else: 
            shutil.copy(olympia_path+"/cont/"+str(j), dst_path+dst_phase[k] + str(j))
        l+=1
        print(str(l)+" images transferred!")
    k+=1

print('Process done!!')
elapsed_time = time.time() - start
print ("elapsed_time:{0}".format(elapsed_time) + "[sec]")
```
